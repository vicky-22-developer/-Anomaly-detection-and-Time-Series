#1. What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

-> Anomaly Detection is a data analysis technique used to identify data points, patterns, or observations that deviate significantly from normal behavior. These unusual observations are called anomalies or outliers and often indicate errors, rare events, fraud, system failures, or novel patterns.

It is widely used in fraud detection, network security, medical diagnosis, industrial monitoring, and finance.

Types of Anomalies
1. Point Anomalies

A point anomaly occurs when a single data instance is significantly different from the rest of the data.

üîπ Example:

In a dataset of daily bank transactions, most amounts are between ‚Çπ500 ‚Äì ‚Çπ20,000, but one transaction of ‚Çπ10,00,000 appears.

That single transaction is a point anomaly.

2. Contextual Anomalies

A contextual anomaly is anomalous only in a specific context (such as time, location, or condition).

It depends on:

Contextual attributes (time, season, location)

Behavioral attributes (measured values)

üîπ Example:

Temperature of 30¬∞C:

Normal in summer

Anomalous in winter

High electricity usage at 3 AM in a household may be anomalous, but normal at 7 PM.

3. Collective Anomalies

A collective anomaly occurs when a group of data points together is anomalous, even if individual points look normal.

üîπ Example:

A sequence of small network requests that individually seem normal but together form a DDoS attack.

Sudden gradual increase in server response time over several minutes.

#2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

->
# 1. Isolation Forest (iForest)

###Approach

- Based on the idea that anomalies are easier to isolate than normal points.

- Uses randomly generated decision trees.

- Anomalies require fewer splits to be isolated ‚Üí shorter path length.

- Model-based and does not rely on distance or density.

###Key Characteristics

- Works well in high-dimensional data

- Efficient for large datasets

- Assumes anomalies are few and different

###Suitable Use Cases

- Fraud detection

- Network intrusion detection

- High-dimensional datasets (e.g., transaction features)

- Real-time anomaly detection.

#2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

###Approach

- Density-based clustering algorithm.

- Points in low-density regions are labeled as anomalies (noise).

- Uses two parameters:

- Œµ (epsilon): neighborhood radius

- MinPts: minimum points to form a dense region

###Key Characteristics

- Identifies clusters of arbitrary shape

- Automatically detects outliers as noise

- Struggles with varying densities and high-dimensional data

###Suitable Use Cases

- Spatial data analysis

- Geographic and image data

- Noise detection in low-dimensional datasets

- When clusters are unknown beforehand.

#3. Local Outlier Factor (LOF)

###Approach

- Density-based local method.

- Compares local density of a point with densities of its neighbors.

- A point is anomalous if its density is significantly lower than its neighbors.

###Key Characteristics

- Detects local anomalies

- Sensitive to choice of k (neighbors)

- Computationally expensive for large datasets

###Suitable Use Cases

- Datasets with varying local densities

- Detecting anomalies near cluster boundaries

- Sensor data and monitoring systems.


#3. What are the key components of a Time Series? Explain each with one example.

-> Key Components of a Time Series

A time series is a sequence of observations recorded over time at regular intervals. Time series data is commonly analyzed in forecasting, economics, finance, weather analysis, and machine learning.

A time series is typically decomposed into the following four key components:

##1. Trend (T)

Definition

The trend represents the long-term movement or overall direction of the data over time (upward, downward, or stable).

Example

- A company‚Äôs annual sales increasing steadily over 10 years due to business growth.

- Sales rising from ‚Çπ10 lakh to ‚Çπ50 lakh over several years.

##2. Seasonality (S)

Definition

Seasonality refers to regular and repeating patterns that occur at fixed intervals (daily, monthly, yearly).

Example

- Ice cream sales peaking every summer and dropping in winter.

- Increased online shopping during festival seasons.

 Repeats at known, fixed periods.

##3. Cyclical Component (C)

Definition

Cyclical variations are long-term fluctuations caused by economic or business cycles, but do not have a fixed period.

Example

- Economic booms and recessions affecting GDP growth or employment rates over several years.

-  Occurs over years, not fixed like seasonality.

##4. Irregular / Random Component (R)

Definition

The irregular component captures random, unpredictable variations caused by unexpected events.

Example

- Sudden drop in airline bookings due to a natural disaster or pandemic.

- Stock market crash due to unexpected news.

- Noise that cannot be predicted easily.

#4.  Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

-> Stationarity in Time Series

Definition

A time series is said to be stationary if its statistical properties do not change over time. In other words:

- Mean is constant over time

- Variance is constant over time

- Autocovariance / autocorrelation depends only on the lag, not on time

Stationarity is important because most time-series models (ARMA, ARIMA) assume the data is stationary.

###How to Test Stationarity:
###1. Visual Inspection

- Plot the time series.

- If you observe trend or seasonality, the series is likely non-stationary.

 Example: A steadily increasing sales series ‚Üí non-stationary.

###2. Rolling Statistics

- Compute rolling mean and rolling variance.

- If they change over time, the series is non-stationary.

###3. Augmented Dickey-Fuller (ADF) Test

- Null hypothesis (H‚ÇÄ): Series has a unit root (non-stationary).

- Alternative hypothesis (H‚ÇÅ): Series is stationary.

###4. KPSS Test (Complementary Test)

- H‚ÇÄ: Series is stationary

- H‚ÇÅ: Series is non-stationary

Using ADF + KPSS together gives more reliable results.

#5.  Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

->
##1. AR (Autoregressive) Model
###Structure

- Uses past values of the series to predict the current value.

- Denoted as AR(p), where p = number of lag observations.

###Application

- Suitable for stationary time series

- When current value depends on previous values

- Example: Predicting temperature based on previous days‚Äô temperatures.

##2. MA (Moving Average) Model
###Structure

- Uses past forecast errors to predict current value.

- Denoted as MA(q), where q = number of lagged errors.


###Application

- Suitable for stationary series

- Captures short-term shock effects

- Example: Modeling stock price fluctuations due to sudden news.

##3. ARIMA (Autoregressive Integrated Moving Average)
###Structure

- Combines AR + differencing + MA

- Denoted as ARIMA(p, d, q):

p = AR order

d = number of differences

q = MA order

###Application

- Used for non-stationary time series

- Widely applied in forecasting

- Example: Sales forecasting with trend.

#6. Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components.


```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

data = pd.read_csv('/content/sample_data/AirPassengers.csv')
data["Month"] = pd.to_datetime(data["Month"])
data.set_index("Month", inplace=True)

# Plot original series
plt.figure()
plt.plot(data["#Passengers"])
plt.title("AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("#Passengers")
plt.show()

# Decompose time series
decomposition = seasonal_decompose(
    data["#Passengers"],
    model="multiplicative",
    period=12
)

decomposition.plot()
plt.show()
```

#7. Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# -------------------------------
# 1. Create / Load Dataset
# -------------------------------
# (Synthetic NYC Taxi‚Äìlike data)
np.random.seed(5)

n_normal = 500
trip_distance = np.random.gamma(shape=2.0, scale=3.0, size=n_normal)
fare_amount = 2.5 + trip_distance * 2.2 + np.random.normal(0, 2.0, n_normal)

# Inject anomalies
n_anomalies = 25
trip_distance_anom = np.random.uniform(0.1, 30, n_anomalies)
fare_amount_anom = np.random.uniform(-10, 150, n_anomalies)

df = pd.DataFrame({
    "trip_distance": np.concatenate([trip_distance, trip_distance_anom]),
    "fare_amount": np.concatenate([fare_amount, fare_amount_anom])
})

# -------------------------------
# 2. Fit Isolation Forest
# -------------------------------
iso_forest = IsolationForest(
    n_estimators=200,
    contamination=0.05,
    random_state=5
)

df["anomaly"] = iso_forest.fit_predict(
    df[["trip_distance", "fare_amount"]]
)

# -------------------------------
# 3. Separate Normal vs Anomalies
# -------------------------------
normal = df[df["anomaly"] == 1]
anomalies = df[df["anomaly"] == -1]

# -------------------------------
# 4. Visualization (2D Scatter)
# -------------------------------
plt.figure(figsize=(8, 6))
plt.scatter(
    normal["trip_distance"],
    normal["fare_amount"],
    alpha=0.6,
    label="Normal Trips"
)

plt.scatter(
    anomalies["trip_distance"],
    anomalies["fare_amount"],
    marker="x",
    s=80,
    label="Anomalies"
)

plt.xlabel("Trip Distance (miles)")
plt.ylabel("Fare Amount ($)")
plt.title("Isolation Forest Anomaly Detection (NYC Taxi Fare Data)")
plt.legend()
plt.show()
```

#8. Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.


```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX

data = pd.read_csv('/content/sample_data/AirPassengers.csv')
data["Month"] = pd.to_datetime(data["Month"])
data.set_index("Month", inplace=True)
data.index.freq = 'MS' # Explicitly set the frequency to Month Start

# Train SARIMA model
model = SARIMAX(data["#Passengers"],order=(1, 1, 1),seasonal_order=(1, 1, 1, 12))
results = model.fit()

# Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean

forecast_index = pd.date_range(start=data.index[-1] + pd.offsets.MonthEnd(1),periods=12,freq="ME")

forecast_series = pd.Series(forecast_mean.values, index=forecast_index)

# Plot actual vs forecast
plt.figure()
plt.plot(data["#Passengers"], label="Actual")
plt.plot(forecast_series, label="Forecast")
plt.title("SARIMA Forecast for Next 12 Months")
plt.xlabel("Year")
plt.ylabel("#Passengers")
plt.legend()
plt.show()
```

#9.  Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

df = pd.read_csv('/content/sample_data/AirPassengers.csv')
df["Month"] = pd.to_datetime(df["Month"])
df.set_index("Month", inplace=True)

# Apply LOF
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
df["Anomaly"] = lof.fit_predict(df[["#Passengers"]])

# Extract anomalies
anomalies = df[df["Anomaly"] == -1]

# Visualization

plt.figure(figsize=(12,6))
plt.plot(df.index, df["#Passengers"], label="#Passengers")
plt.scatter(anomalies.index, anomalies["#Passengers"], marker="x", label="Anomalies")
plt.title("LOF Anomaly Detection on Airline Passengers Data")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend()
plt.show()
```
